Another example. Here we have a hist chart that shows how often hostels within a certain price rage occure. What's interesting about this example is that just using the raw data we get a histogram which won't be very useful. Due to outliers our histogram gets messed up. A data scientist also has to deal with these sort of things and find solutions in such a way the data becomes meaningful and insightful again. In our case we just simply remove the two records which have an absurtly high price. In other cases you may want to do something else to the data or use a different visual that's better suited to represent the data which has outliers. 


```python
high_price_filter = hostels['price.from'] < 10**5
hostels['price.from'][high_price_filter].plot(kind='hist', bins=14)
```
